```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")          
sns.set_palette("husl")                   
plt.rcParams['figure.figsize'] = (10, 6)  
plt.rcParams['axes.titlesize'] = 14       
df = pd.read_csv('nfhs5_india_district_health_data.csv')
print("Libraries imported and visualization environment configured successfully!")
```

```python
print(f"Rows: {df.shape[0]} | Columns: {df.shape[1]}")
print(df.columns.tolist())
df.info()
display(df.head())
df.describe(include='all')
```

```python
cols_for_heatmap = ['Women_Secondary_Education_pct', 'Child_Marriage_Women_pct','Total_Fertility_Rate', 'Contraceptive_Prevalence_Rate_pct',
'Households_Improved_Sanitation_pct', 'Households_Electricity_pct','Institutional_Births_pct','Children_Under5_Stunted_pct',
'Anaemia_Women_15_49yrs_pct','Anaemia_Children_6_59months_pct']
corr_matrix = df[cols_for_heatmap].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1, square=True, linewidths=0.5)      
plt.title('Correlation Heatmap of Key Health & Socio-Economic Indicators', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.regplot(data=df, x='Women_Secondary_Education_pct', y='Children_Under5_Stunted_pct', ax=axes[0],scatter_kws={'alpha':0.5, 'color':'steelblue'}, 
line_kws={'color':'red', 'linewidth':2})
axes[0].set_title("Impact of Women's Education on Child Stunting", fontsize=14, pad=15)
axes[0].set_xlabel("Women with Secondary Education (%)")
axes[0].set_ylabel("Children Under 5 Stunted (%)")
sns.regplot(data=df, x='Women_Secondary_Education_pct', y='Child_Marriage_Women_pct', ax=axes[1],scatter_kws={'alpha':0.5, 'color':'darkorchid'}, 
line_kws={'color':'red', 'linewidth':2})
axes[1].set_title("Impact of Women's Education on Child Marriage", fontsize=14, pad=15)
axes[1].set_xlabel("Women with Secondary Education (%)")
axes[1].set_ylabel("Child Marriage (%)")
plt.tight_layout()
plt.show()
```

```python
top_5_states = df.groupby('State_UT')['Anaemia_Women_15_49yrs_pct'].mean().nlargest(5).index
df_top5 = df[df['State_UT'].isin(top_5_states)]
df_melted = df_top5.melt(id_vars='State_UT', value_vars=['Anaemia_Women_15_49yrs_pct', 'Anaemia_Men_15_49yrs_pct'],
var_name='Gender', value_name='Anaemia_Prevalence')
df_melted['Gender'] = df_melted['Gender'].replace({'Anaemia_Women_15_49yrs_pct': 'Women (15-49 yrs)','Anaemia_Men_15_49yrs_pct': 'Men (15-49 yrs)'})
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x='State_UT', y='Anaemia_Prevalence', hue='Gender', palette='Set2', errorbar=None)
plt.title('Gender Gap in Anaemia Prevalence (Top 5 Most Affected States)', fontsize=15, pad=15)
plt.xlabel('State / Union Territory', fontsize=12)
plt.ylabel('Average Anaemia Prevalence (%)', fontsize=12)
plt.legend(title='Demographic', bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend outside the plot
plt.tight_layout()
plt.show()
```

```python
pairplot_cols = ['Women_Secondary_Education_pct', 'Total_Fertility_Rate', 'Child_Marriage_Women_pct','Children_Under5_Stunted_pct']
g = sns.pairplot(df[pairplot_cols], corner=True, diag_kind='kde',plot_kws={'alpha': 0.4, 'color': 'teal', 'edgecolor': 'w'},
diag_kws={'color': 'teal', 'fill': True})
g.fig.suptitle("Pairwise Relationships: Education, Fertility, and Health", y=1.05, fontsize=16)
plt.show()
```

```python
top_5_marriage_states = df.groupby('State_UT')['Child_Marriage_Women_pct'].mean().nlargest(5).index
df_violin = df[df['State_UT'].isin(top_5_marriage_states)]
plt.figure(figsize=(12, 7))
sns.violinplot(data=df_violin, x='State_UT', y='Child_Marriage_Women_pct', hue='State_UT', palette='magma', inner='quartile', 
linewidth=1.5, legend=False)
plt.title('Distribution of Child Marriage Rates by State (Top 5 Highest)', fontsize=15, pad=15)
plt.xlabel('State / Union Territory', fontsize=12)
plt.ylabel('Child Marriage (%)', fontsize=12)
national_avg = df['Child_Marriage_Women_pct'].mean()
plt.axhline(national_avg, color='black', linestyle='--', label=f'National Average ({national_avg:.1f}%)')
plt.xticks(rotation=45, ha='right')
plt.legend() 
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Women_Secondary_Education_pct', y='Total_Fertility_Rate',scatter_kws={'alpha':0.4, 'color':'forestgreen'},
line_kws={'color':'red', 'linewidth':2})
plt.axhline(2.1, color='black', linestyle='--', label='Replacement Level (2.1)')
plt.title("The Impact of Education on Fertility Rates", fontsize=15, pad=15)
plt.xlabel("Women with Secondary Education (%)")
plt.ylabel("Total Fertility Rate (Children per Woman)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
top_states = df['State_UT'].value_counts().nlargest(10).index
df_subset = df[df['State_UT'].isin(top_states)]
sns.scatterplot(data=df_subset, x='Women_Secondary_Education_pct', y='Children_Under5_Stunted_pct',hue='State_UT', size='Households_Surveyed', 
sizes=(20, 400), alpha=0.6, palette='viridis')
plt.title("4-Dimensional Analysis: Education, Stunting, Geography, and Scale", fontsize=16, pad=20)
plt.xlabel("Women with Secondary Education (%)")
plt.ylabel("Children Under 5 Stunted (%)")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='State & Survey Scale')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
unmet_need_state = df.groupby('State_UT')['Unmet_Need_Family_Planning_pct'].mean().nlargest(10).reset_index()
sns.barplot(data=unmet_need_state, x='Unmet_Need_Family_Planning_pct', y='State_UT', hue='State_UT', palette='Reds_r',    
legend=False)
plt.title("Top 10 States with Highest Unmet Need for Family Planning", fontsize=15, pad=15)
plt.xlabel("Average Unmet Need (%)", fontsize=12)
plt.ylabel("State / Union Territory", fontsize=12)
for index, value in enumerate(unmet_need_state['Unmet_Need_Family_Planning_pct']):
    plt.text(value + 0.2, index, f'{value:.1f}%', va='center', fontsize=10)
plt.tight_layout()
plt.show()
```
